## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
cur.execute("SELECT COUNT(*) FROM CHICAGO_CRIME_DATA")
row_count = cur.fetchone()[0]  # Fetch the count value from the first row
print("Number of rows in CHICAGO_CRIME_DATA:", row_count)
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
query = """
SELECT community_area_name, per_capita_income
FROM CENSUS_DATA
WHERE per_capita_income < 11000;
"""
cur.execute(query)
communities = cur.fetchall()
print(communities)
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)



```python
query = """
SELECT CASE_NUMBER, DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE DESCRIPTION like '%minor%'
"""
cur.execute(query)
crimeNo = cur.fetchall()
print(crimeNo)
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
query = """
SELECT PRIMARY_TYPE, DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION like '%child%'
"""
cur.execute(query)
Kidnapp = cur.fetchall()
print(Kidnapp)
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
query = """
SELECT PRIMARY_TYPE, LOCATION_DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE LOCATION_DESCRIPTION like '%school%'
"""
cur.execute(query)
schools = cur.fetchall()
print(schools)
```

### Problem 6

##### List the average safety score for each type of school.


```python
query = """
SELECT "ELEMENTARY, MIDDLE, OR HIGH SCHOOL", AVG("SAFETY_SCORE")
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
group by "ELEMENTARY, MIDDLE, OR HIGH SCHOOL"
"""
cur.execute(query)
safety = cur.fetchall()
print(safety)
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line 


```python
query = """
SELECT "COMMUNITY_AREA_NAME", "PERCENT_HOUSEHOLDS_BELOW_POVERTY" 
from CENSUS_DATA 
order by "PERCENT_HOUSEHOLDS_BELOW_POVERTY" desc 
"""
cur.execute(query)
community = cur.fetchall()
print(community)
```

### Problem 8

##### Which community area is most crime prone?


```python
query = """
SELECT COMMUNITY_AREA_NUMBER, count(COMMUNITY_AREA_NUMBER) 
from CHICAGO_CRIME_DATA 
ORDER BY (COMMUNITY_AREA_NUMBER)  DESC LIMIT 1
"""
cur.execute(query)
pron = cur.fetchall()
print(pron)
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
query = """
SELECT "COMMUNITY_AREA_NAME", "HARDSHIP_INDEX"
FROM CENSUS_DATA
WHERE "HARDSHIP_INDEX" = (
    SELECT MAX("HARDSHIP_INDEX") FROM CENSUS_DATA
);
"""
cur.execute(query)
hardship = cur.fetchall()
print(hardship)
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
query = """
SELECT "COMMUNITY_AREA_NAME"
FROM CENSUS_DATA 
WHERE "COMMUNITY_AREA_NUMBER" IN (
    SELECT COUNT("COMMUNITY_AREA_NUMBER")
    from CHICAGO_CRIME_DATA 
    GROUP BY COMMUNITY_AREA_NUMBER
    ORDER BY ("COMMUNITY_AREA_NUMBER")  DESC LIMIT 1

);
"""

cur.execute(query)
ComCrimes = cur.fetchall()
print(ComCrimes)
```
